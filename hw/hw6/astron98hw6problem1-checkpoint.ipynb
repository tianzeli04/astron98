```python
y1=np.random.rand(40)
print(y1)
x1=np.arange(1, 41, 1)
print(x1)
y2=np.random.rand(40)
print(y2)
```

```python
fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x1, y1, color='orange', linewidth=10) 
axs[0].plot(x1, y2, color='red') 
axs[1].scatter(x1, y2, color='magenta', marker='d')    
```
