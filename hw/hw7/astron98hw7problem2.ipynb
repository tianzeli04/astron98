```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('planets') 
df
```

```python
plt.figure()
plt.scatter(df['orbital_period'], df['mass'], label='data')
plt.loglog()
plt.title('Mass vs Orbital Period')
plt.xlabel('Orbital Period')
plt.ylabel("Mass")
plt.legend()
plt.show()

#with individual points colored by discovery method
```

```python
plot=sns.scatterplot(x=df['orbital_period'], y=df['mass'], hue=df['method'])
plot.set(xscale='log', yscale='log')
plot.set_title('Mass vs Orbital Period')
plot.set_xlabel('Orbital Period')
plot.set_ylabel("Mass")
```

```python
sns.histplot(df['number'], x=df['year'], hue=df['method'])
```
